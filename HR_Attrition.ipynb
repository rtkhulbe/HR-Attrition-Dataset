```python

# plot attrtion based on marital status and gender

plt.figure(figsize = (12,8))
df_attrition = df[df["Attrition"] == "Yes"]

df_heatmap = df_attrition.pivot_table(index="MaritalStatus", columns="Gender", values="EmployeeNumber", aggfunc="count")

plt.figure(figsize=(8,6))
sns.heatmap(df_heatmap, annot=True, cmap="coolwarm", fmt="d")
plt.title("Attrition by Marital Status & Gender (Heatmap)")
plt.xlabel("Gender")
plt.ylabel("Marital Status")
plt.show()


```

Key Observation:

Males consistently have higher attrition counts than females across all marital statuses.

The Single-Male group stands out with the highest attrition (73), followed by Single-Female (47).

The lowest attrition is seen among Divorced-Female employees (9).

```python
# Atrrition of the employees that leave the organization within one year

# Assuming 'YearsAtCompany' is the column for tenure
df_attrition = df[df['YearsAtCompany'] <= 1]  # Select employees with <= 1 year

# Group the data to create funnel stages
funnel_data = {
    "stage": ["Total Employees", "Joined Last Year", "Employees Leaving < 1 Year", "Retained Employees"],
    "count": [df.shape[0], df[df['YearsAtCompany'] <= 1].shape[0], df_attrition.shape[0], df.shape[0] - df_attrition.shape[0]]
}

# Create funnel chart
# Sort the funnel_data by 'count' in ascending order before plotting

funnel_df = pd.DataFrame(funnel_data)
funnel_df = funnel_df.sort_values(by="count", ascending=False)

fig = px.funnel(funnel_df, x="count", y="stage", title="Employee Attrition Funnel (Leaving within 1 Year)")
fig.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Age', y='MonthlyIncome', color='purple')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.title('Scatter Plot of Age vs. Monthly Income')

# Show the plot
plt.show()
```

```python
# Aggregate to get mean PercentSalaryHike for each YearsAtCompany
years_salary = df.groupby('YearsAtCompany')['PercentSalaryHike'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(years_salary['YearsAtCompany'], years_salary['PercentSalaryHike'], marker='o', linestyle='-', color='purple')

plt.xlabel('Years at Company')
plt.ylabel('Average Percent Salary Hike')
plt.title('Years at Company vs. Average Salary Hike')
plt.grid(True)
plt.show()
```

Key Observations:

For employees with fewer years at the company (0–20 years), the average salary hike percentage generally fluctuates between 14% and 16%, with occasional peaks above 16%.

After 20 years, the data becomes more volatile, with sharp rises and drops in the average salary hike percentage.

The highest average salary hikes (above 17%) occur sporadically at several points, notably around 27 and 36 years.

There are also notable drops, with the lowest average salary hike (around 11%) occurring at approximately 34 years.

Overall, while the average salary hike remains relatively stable for most of the tenure range, there is increased variability for employees with longer tenure.
